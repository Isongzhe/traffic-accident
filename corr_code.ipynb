```python
#基本套件
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False    
#讀檔，用utf-8的格式開
df = pd.read_csv('./dataset/processed_data/97-111 A1事件預測資料集.csv', encoding='utf-8')
#將df['統計期']轉成df的index，並把index的名字移除
df.set_index('統計期', inplace=True) 
df.index.name = None
```

```python
#畫關聯矩陣

plt.figure(figsize=(12,10), dpi= 80)
#設定遮罩，遮住上半部
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
sns.heatmap(df.corr(), 
            xticklabels=df.corr().columns, 
            yticklabels=df.corr().columns, 
            cmap='coolwarm', 
            center=0, 
            annot=True, 
            mask=mask)
plt.title('預測參數的關聯矩陣', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
plt.savefig('./dataset/graph/預測參數的關聯矩陣.png')
```
