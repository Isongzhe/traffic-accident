民國年換西元年

```python
name ='事故件數(件)'
folder_path = './dataset/processed_data/97-111 道路交通{}.csv'.format(name)
df = pd.read_csv(folder_path)
processed_dates = []  # 儲存處理後的日期字串

for date_str in df['統計期']:
    parts = str(date_str)  # 將 date_str 轉換為字串
    parts = parts.split('年 ')
    year = int(parts[0])+ 1911
    date_str = str(year) + '年' +parts[1]
    processed_dates.append(date_str)
processed_dates = pd.to_datetime(processed_dates, format='%Y年%M月').strftime('%Y-%M')
df['統計期'] = processed_dates
```

```python
six_city = df.iloc[:, 0:8] 
```

箱型圖

```python
# 設定seaborn的格式:style="dark"背景為暗色, palette="pastel"設定顏色調色板為柔和色系
sns.set(style="dark", palette="pastel", font='Microsoft JhengHei') #font設定讓seaborn使用matplotlib的中文

def box_figure(items):
    # 畫圖
    plt.figure(figsize=(8, 6)) 
    plot = sns.boxplot(data=six_city)
    plt.ylabel('{}'.format(items), fontsize=12)
    plt.title('97-111年 六都交通{}_箱型圖'.format(items), fontsize = 18, weight ='bold') #weight ='bold'為粗體
    # plt.tick_params(axis='both', which='major', labelsize = 10) #設定刻度:axis='both'為兩軸, which='major'為主軸
    plt.grid()
    file_name = "./dataset/graph/97-111年_六都交通{}箱型圖.png".format(items)
    plt.savefig(file_name, bbox_inches='tight') # bbox_inches='tight' 參數將會嘗試將圖形緊密地貼合在圖片區域內
    plt.show()
box_figure(name)
```

```python
#用來產生圖的，方便之後迴圈輸出
def generate_graph(i):
    xticks = df.index[::12].tolist()
    plt.figure(figsize=(8, 6))
    plt.plot(df[i])
    plt.title('97-111年 {}_道路交通事故件數(件)'.format(i))
    plt.xlabel('時間(月)')
    plt.xticks(xticks)
    plt.xticks(rotation=60) 
    plt.ylabel('事故件數(件)')
    plt.savefig('./dataset/graph/97-111_{}交通事故.png'.format(i))

for _,x in enumerate(df.columns):
    generate_graph(x)
```

```python
import plotly.graph_objects as go
import pandas as pd
import os

# df = pd.read_csv('./dataset/processed_data/道路交通/97-111 每百萬人道路交通事故件數(件).csv', index_col=0).iloc[:, 1:7] 
# df_2 = pd.read_csv('./dataset/processed_data/道路交通/97-111 道路交通事故件數(件).csv', index_col=0).iloc[:, 1:7] 

# num = df.columns
# 指定特定的資料夾路徑
folder_path = './dataset/processed_data/道路交通/'
file_names = ['97-111 道路交通事故件數(件).csv', 
              '97-111 每百萬人道路交通事故件數(件).csv',
              '97-111 道路交通死亡人數(人).csv',
              '97-111 每百萬人道路交通死亡人數(人).csv',
              '97-111 道路交通受傷人數(人).csv',
              '97-111 每百萬人道路交通受傷人數(人).csv']

def boxplot_savefig():
    
    fig = go.Figure()
    first_iteration = True  # 用來判斷是否是第一次迴圈
    for file_name in file_names:
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path, index_col=0).iloc[:, 1:7] #設定為六都數據
        
        for column in df.columns:
            if first_iteration:
                fig.add_trace(go.Box(y=df[column], name=column, visible=True))
            else:
                fig.add_trace(go.Box(y=df[column], name=column, visible=False))
        
        first_iteration = False  # 將標誌設為 False，表示後續迴圈將跑else:

    num = df.columns
    # 建立按钮列表
    buttons = [
        dict(
            label="事故件數(件)",
            method="update",
            args=[{"visible": [True]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)}, 
                {"title": "六都道路交通事故件數(件)的箱型圖", "annotations": []}]
        ),
        dict(
            label="每百萬人_事故件數(件)",
            method="update",
            args=[{"visible": [False]*len(num)+[True]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)}, 
                {"title": "每百萬人_六都道路交通事故件數(件)的箱型圖", "annotations": []}]
        ),
        dict(
            label="死亡人數(人)",
            method="update",
            args=[{"visible": [False]*len(num)+[False]*len(num)+[True]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)}, 
                {"title": "六都道路交通死亡人數(人)的箱型圖", "annotations": []}]
        ),
        dict(
            label="每百萬人_死亡人數(人)",
            method="update",
            args=[{"visible": [False]*len(num)+[False]*len(num)+[False]*len(num)+[True]*len(num)+[False]*len(num)+[False]*len(num)}, 
                {"title": "每百萬人_六都道路交通死亡人數(人)的箱型圖", "annotations": []}]
        ),
        dict(
            label="受傷人數(人)",
            method="update",
            args=[{"visible": [False]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)+[True]*len(num)+[False]*len(num)}, 
                {"title": "六都道路交通受傷人數(人)的箱型圖", "annotations": []}]
        ),
        dict(
            label="每百萬人_受傷人數(人)",
            method="update",
            args=[{"visible": [False]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)+[False]*len(num)+[True]*len(num)}, 
                {"title": "每百萬人_六都道路交通受傷人數(人)的箱型圖", "annotations": []}]
        )]
    fig.update_layout(title='六都交通事故數據盒方圖', updatemenus = [dict(buttons=buttons)])
    fig.write_html("./dataset/graph/六都交通事故數據盒方圖.html")
    return fig

boxplot_savefig()
```
